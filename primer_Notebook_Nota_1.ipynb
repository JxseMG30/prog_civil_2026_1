# 1. Problema de ingeniería sencillo

Supongamos un problema típico en **ingeniería civil o agrícola**:

**Calcular el volumen de agua almacenado en un tanque rectangular.**

La fórmula física es:

[
V = largo \times ancho \times altura
]

donde:

* largo (m)
* ancho (m)
* altura (m)

---

# 2. Definir ENTRADAS – PROCESO – SALIDAS

### Entradas (datos que proporciona el usuario)

* largo del tanque (m)
* ancho del tanque (m)
* altura del tanque (m)

### Proceso (cálculo que realiza el algoritmo)

1. Leer los valores de largo, ancho y altura.
2. Aplicar la fórmula del volumen.

[
volumen = largo \times ancho \times altura
]

### Salida (resultado)

* volumen del tanque en **metros cúbicos (m³)**

---

# 3. Algoritmo (explicación lógica)

1. Iniciar programa.
2. Solicitar al usuario el largo del tanque.
3. Solicitar el ancho del tanque.
4. Solicitar la altura del tanque.
5. Calcular el volumen.
6. Mostrar el volumen.
7. Finalizar programa.

---

# 4. Pseudocódigo

```text
Inicio

leer largo
leer ancho
leer altura

volumen = largo * ancho * altura

mostrar volumen

Fin
```

---

# 5. Código Python (con variables y operadores)

```python
# Programa para calcular el volumen de un tanque rectangular

# Entradas
largo = float(input("Ingrese el largo del tanque (m): "))
ancho = float(input("Ingrese el ancho del tanque (m): "))
altura = float(input("Ingrese la altura del tanque (m): "))

# Proceso
volumen = largo * ancho * altura

# Salida
print("El volumen del tanque es:", volumen, "m^3")
```

---

# 6. Ejemplo de ejecución

Entrada del usuario:

```
largo = 5
ancho = 3
altura = 2
```

Resultado:

```
El volumen del tanque es: 30 m^3
```
